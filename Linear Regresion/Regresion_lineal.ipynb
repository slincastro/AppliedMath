### Regresion Lineal Con Gradiente Descendiente

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo (Tamaño de la casa en m² vs Precio en miles de dólares)

X = np.array([50, 80, 100, 120])  # Tamaño de la casa
y = np.array([150, 200, 250, 300])  # Precio de la casa

# Normalización de datos (opcional pero recomendable)
X = (X - np.mean(X)) / np.std(X)

# Parámetros iniciales
theta_0 = 0
theta_1 = 0
alpha = 0.01  # Tasa de aprendizaje
epochs = 1000  # Número de iteraciones

# Implementación de Descenso de Gradiente

m = len(X)  # Número de ejemplos
for _ in range(epochs):
    y_pred = theta_0 + theta_1 * X  # Predicción
    error = y_pred - y  # Error
    
    # Cálculo de los gradientes
    theta_0 -= alpha * (1/m) * np.sum(error)
    theta_1 -= alpha * (1/m) * np.sum(error * X)

# Predicciones finales
y_pred = theta_0 + theta_1 * X

# Visualización
plt.scatter(X, y, color="blue", label="Datos reales")
plt.plot(X, y_pred, color="red", label="Regresión Lineal")
plt.xlabel("Tamaño de la casa (normalizado)")
plt.ylabel("Precio de la casa (miles de $)")
plt.legend()
plt.show()

# Valores finales de theta
print(f"Intercepto (theta_0): {theta_0}")
print(f"Pendiente (theta_1): {theta_1}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos
X = np.array([50, 80, 100, 120]).reshape(-1, 1)  # Debe ser una matriz 2D
y = np.array([150, 200, 250, 300])

# Crear modelo
model = LinearRegression()
model.fit(X, y)  # Entrenar el modelo

# Hacer predicciones
y_pred = model.predict(X)

# Visualización
plt.scatter(X, y, color="blue", label="Datos reales")
plt.plot(X, y_pred, color="red", label="Regresión Lineal")
plt.xlabel("Tamaño de la casa (m²)")
plt.ylabel("Precio de la casa (miles de $)")
plt.legend()
plt.show()

# Parámetros del modelo
print(f"Intercepto (theta_0): {model.intercept_}")
print(f"Pendiente (theta_1): {model.coef_[0]}")
```
