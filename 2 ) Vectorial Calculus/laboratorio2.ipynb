# 8. Analizando el Test Accuracy Vs el porcentaje de muestra 
### Para encontrar los mejores porcentajes de training y test con diferentes Solvers

[Indice](#indice)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

solvers = ['lbfgs', 'sgd', 'adam']
hiden_layers =(2,4)


def train_graph(solver):
    train_sizes = np.arange(30, 100, 1)  # De 30% a 90% en pasos de 10%
    accuracies = []


    for train_size in train_sizes:
        X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size/100, random_state=1)

        #clf = MLPClassifier(solver=solver, alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1, max_iter=3000)
        clf = MLPClassifier(solver=solver, alpha=1e-5,activation="identity", hidden_layer_sizes=hiden_layers, random_state=1, max_iter=3000)

        #hidden_layer_sizes=(20, 10, 5) 
        clf.fit(X_train, y_train)

        y_pred = clf.predict(X_test)

        accuracy = accuracy_score(y_test, y_pred)
        accuracies.append(accuracy)



    max_index = np.argmax(accuracies)
    best_train_size = train_sizes[max_index]
    best_accuracy = accuracies[max_index]

    plt.figure(figsize=(8, 5))
    plt.plot(train_sizes, accuracies, marker='o', linestyle='-', color='blue', label="Test Accuracy")

    plt.axvline(x=best_train_size, color='red', linestyle='--', label=f"Max Accuracy ({best_accuracy:.2f})")

    plt.xlabel("Training Data Percentage")
    plt.ylabel("Test Accuracy")
    plt.title("MLP Accuracy vs. Training Data Size (Iris Dataset) Con el solver: " + solver)
    plt.legend()
    plt.grid()

    plt.show()

    print(f"Best Accuracy: {best_accuracy:.4f} at Training Size: {best_train_size}%")

for solver in solvers:
    print(f"Solver: {solver}")
    train_graph(solver)
    print("\n\n")
```

# A. Epochs (numero de iteraciones) vs Accuracy (precision)

Calculando el acuracy vs presicion en diferentes configuraciones de Hiden layers

```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

warnings.filterwarnings("ignore")

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.86, random_state=1)

def detect_change_points(accuracy_list, threshold=0.1):
    """Detects points where accuracy changes abruptly."""
    change_points = []
    for i in range(1, len(accuracy_list)):
        if abs(accuracy_list[i] - accuracy_list[i-1]) > threshold:
            change_points.append(i)  
    return change_points

def plot_mlp_variations(rows=5, cols=5, max_epochs=50, change_threshold=0.1):
    fig, axes = plt.subplots(rows, cols, figsize=(20, 20))
    fig.suptitle("MLP Test Accuracy sobre algunas epocas variando Configiraciones de Capas Ocultas (Different Configurations)", fontsize=16)

    for i in range(rows):
        for j in range(cols):
            ax = axes[i, j]

            hidden_layers = tuple(np.random.randint(1, 10, size=np.random.randint(1, 3)))  # 1 to 2 layers
            random_seed = np.random.randint(0, 100)

            accuracy_test = []
            epoch_range = range(1, max_epochs)

            clf = MLPClassifier(hidden_layer_sizes=hidden_layers, solver='adam', activation="relu",
                                max_iter=1, warm_start=True, random_state=random_seed)

            for epoch in epoch_range:
                clf.max_iter = epoch
                clf.fit(X_train, y_train)
                y_pred = clf.predict(X_test)
                accuracy = accuracy_score(y_test, y_pred)
                accuracy_test.append(accuracy)

            change_points = detect_change_points(accuracy_test, threshold=change_threshold)

            ax.plot(epoch_range, accuracy_test, marker='o', linestyle='-', color='blue', label="Test Accuracy")

            for cp in change_points:
                ax.axvline(x=cp, color='red', linestyle='--', alpha=0.7, label="Abrupt Change")

            ax.set_title(f"Layers: {str(hidden_layers)} ", fontsize=10)
            ax.set_xlabel("Epochs", fontsize=8)
            ax.set_ylabel("Test Accuracy", fontsize=8)
            ax.grid(True)

    plt.tight_layout(rect=[0, 0, 1, 0.96])  
    plt.show()

plot_mlp_variations()
```

## Usando la mejor y mas Eficiente Encontrada

```python
import warnings
warnings.filterwarnings("ignore")


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.54, random_state=1)


accuracy_test = []
losses = []
epoch_range = range(1, 100)
for epoch in epoch_range:  # De 1 a 50 epochs

    clf = MLPClassifier(hidden_layer_sizes=hiden_layers , solver='adam',activation="identity", max_iter=epoch, warm_start=True, random_state=1)
    
    clf.fit(X_train, y_train)  

    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_test.append(accuracy)

    losses.append(clf.loss_)



plt.figure(figsize=(8, 4))
plt.plot(epoch_range, accuracy_test, marker='o', linestyle='-', color='blue', label="Test Accuracy")
#plt.axvline(x=best_train_size, color='red', linestyle='--', label=f"Max Accuracy ({best_accuracy:.2f})")

plt.xlabel("Epochs")
plt.ylabel("Test Accuracy")
plt.title("Test Accuracy Over Epochs (MLPClassifier)")
plt.grid()
plt.show()

```

### Sobre los porcentajes de [Entrenamiento](#8-analizando-el-test-accuracy-vs-el-porcentaje-de-muestra) 

Solver: lbfgs (Optimización Cuasi-Newton)
El modelo alcanza una precisión máxima del 100% con un tamaño de entrenamiento de 52%.
Antes de alcanzar el 100%, la precisión muestra cierta inestabilidad en los valores entre 30% y 50% de datos de entrenamiento.
A partir del 52%, la precisión se estabiliza en 1.0, lo que indica que este solver converge rápido y de manera eficiente para este dataset.

Solver: sgd (Gradiente Descendente Estocástico)
Muestra un comportamiento más fluctuante en la precisión antes de alcanzar el 100%.
La precisión sigue variando significativamente hasta aproximadamente 80% de los datos de entrenamiento, donde finalmente alcanza el 100%.
Esto sugiere que el solver sgd requiere más datos para converger a una solución estable.

Solver: adam (Optimización Adaptativa)
La precisión comienza alta, pero disminuye gradualmente a medida que aumenta el tamaño del conjunto de entrenamiento.
A diferencia de lbfgs, que estabiliza temprano, y sgd, que fluctúa, adam parece no converger de manera estable antes del 86% del conjunto de entrenamiento.
Finalmente, a partir del 86%, logra alcanzar una precisión del 100%.

El experimento muestra que la elección de las capas ocultas impacta fuertemente la precisión y estabilidad del modelo. Mientras algunas configuraciones alcanzan alta precisión rápidamente, 
otras presentan oscilaciones o requieren más iteraciones. Esto refuerza la importancia de ajustar correctamente la arquitectura de la red para obtener modelos eficientes y bien generalizados.

El modelo necesita al menos 20-30 épocas para empezar a aprender (período inicial de aprendizaje lento).
