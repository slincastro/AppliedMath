```python
#  Sistemas Lineales, Espacios Vectoriales, Base, Rango y Transformaciones Lineales

## 1锔 **Sistemas Lineales: Soluci贸n mediante Matrices**
### Ejemplo: Resolver un sistema de ecuaciones lineales

import numpy as np
import matplotlib.pyplot as plt

# Definimos la matriz de coeficientes y el vector de constantes
A = np.array([[2, 1], [-1, 1]])
b = np.array([4, 1])

# Resolvemos el sistema Ax = b
x = np.linalg.solve(A, b)
print(f"Soluci贸n: x = {x}")


## 2锔 **Espacios Vectoriales y Bases**
### Ejemplo: Verificar si un conjunto de vectores forma una base

vectors = np.array([[1, 0], [0, 1]])
rank = np.linalg.matrix_rank(vectors)
print(f"Rango: {rank} - Base v谩lida" if rank == 2 else "No es una base")


## 3锔 **Rango de una Matriz**
### Ejemplo: C谩lculo del rango y representaci贸n gr谩fica

A = np.array([[1, 2], [2, 4]])
rango = np.linalg.matrix_rank(A)
print(f"Rango de la matriz: {rango}")


## 4锔 **Transformaciones Lineales**
### Ejemplo: Rotaci贸n de un vector en el plano

theta = np.pi / 4  # ngulo de rotaci贸n (45掳)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])
vector = np.array([1, 0])
rotated_vector = rotation_matrix @ vector
print(f"Vector rotado: {rotated_vector}")


##  **Visualizaci贸n**

plt.figure(figsize=(6,6))
plt.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='r', label='Original')
plt.quiver(0, 0, *rotated_vector, angles='xy', scale_units='xy', scale=1, color='b', label='Rotado')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.gca().set_aspect('equal')
plt.legend()
plt.title('Transformaci贸n Lineal - Rotaci贸n')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt

# Definir matriz aumentada
A = Matrix([[1, 1, 1, 6],
            [2, 3, 1, 14],
            [0, 1, -1, -2]])

# Calcular RREF
rref_matrix, _ = A.rref()
print("Forma RREF:\n", rref_matrix)

# Visualizaci贸n gr谩fica de pivotes
plt.imshow(np.array(rref_matrix).astype(float), cmap='Blues', aspect='auto')
plt.title('Matriz en RREF (Pivotes resaltados)')
plt.colorbar(label='Valor')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import Matrix

# Definir matriz de vectores (como filas)
vectores = np.array([[1, 2, 3],
                     [2, 4, 6],
                     [1, 1, 1]])

# Calcular el rango usando NumPy
rank = np.linalg.matrix_rank(vectores)
print(f"Rango de los vectores: {rank}")

# Calcular RREF usando SymPy
rref_matrix, pivots = Matrix(vectores).rref()
print("\nForma RREF:\n", rref_matrix)

# Visualizaci贸n de vectores en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = ['r', 'g', 'b']
labels = ['v1', 'v2', 'v3']

for i, vec in enumerate(vectores):
    ax.quiver(0, 0, 0, vec[0], vec[1], vec[2], color=colors[i], label=labels[i])

ax.set_xlim([0, 3])
ax.set_ylim([0, 5])
ax.set_zlim([0, 7])
ax.set_title('Visualizaci贸n de Vectores')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de rotaci贸n 45掳
theta = np.pi / 4
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta),  np.cos(theta)]])
vector = np.array([1, 0])
rotated_vector = rotation_matrix @ vector

print(f"Vector rotado: {rotated_vector}")

plt.quiver(0, 0, *vector, angles='xy', scale_units='xy', scale=1, color='r', label='Original')
plt.quiver(0, 0, *rotated_vector, angles='xy', scale_units='xy', scale=1, color='b', label='Rotado')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.gca().set_aspect('equal')
plt.grid(True)
plt.legend()
plt.title('Rotaci贸n en 2D')
plt.show()

scale_matrix = np.array([[2, 0],
                         [0, 3]])
vector = np.array([1, 1])
scaled_vector = scale_matrix @ vector

print(f"Vector escalado: {scaled_vector}")

reflection_matrix = np.array([[-1, 0],
                              [ 0, 1]])
vector = np.array([2, 3])
reflected_vector = reflection_matrix @ vector

print(f"Vector reflejado: {reflected_vector}")
```

```python
import numpy as np

# Definir vectores
v1 = np.array([1, 0])
v2 = np.array([0, 1])

# Producto punto
dot_product = np.dot(v1, v2)
print(f"Producto punto: {dot_product}")

# Verificar ortogonalidad
if dot_product == 0:
    print("Los vectores son ortogonales.")
else:
    print("Los vectores NO son ortogonales.")

# Definir vectores
u1 = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
u2 = np.array([-1/np.sqrt(2), 1/np.sqrt(2)])

# Verificar norma
norm_u1 = np.linalg.norm(u1)
norm_u2 = np.linalg.norm(u2)

# Verificar ortogonalidad
dot = np.dot(u1, u2)

print(f"Norma de u1: {norm_u1}")
print(f"Norma de u2: {norm_u2}")
print(f"Producto punto: {dot}")

if np.isclose(norm_u1, 1) and np.isclose(norm_u2, 1) and np.isclose(dot, 0):
    print("Los vectores forman un conjunto ortonormal.")
else:
    print("Los vectores NO son ortonormales.")
```
